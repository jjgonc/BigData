## Tratamento dos Dados

#### **Transformar `period` em dados numéricos**

```python
# Dicionário para mapear o período número ao valor da string
period_map = {
    'M01': 1,
    'M02': 2,
    'M03': 3,
    'M04': 4,
    'M05': 5,
    'M06': 6,
    'M07': 7,
    'M08': 8,
    'M09': 9,
    'M10': 10,
    'M11': 11,
    'M12': 12,
    'M13:': 13
}

data_csv['period'] = data_csv["period"].replace(period_map)
```

#### **Remover a coluna `footnote_codes`**

```python
data_csv.drop(columns=["footnote_codes"], inplace=True)
```
